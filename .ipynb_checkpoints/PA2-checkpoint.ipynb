```python
x=0
age=int(input("What is the Subject's Age (Type 0 to quit): "))
while age>x:
    if age>=1 and age<=17:
        print ("This person is a minor")
    if age>=18 and age<=64:
        print ("This person is an adult")
    if age>=65:
        print ("This person is a senior citizen")
    age=int(input("What is the Subject's Age (Type 0 to quit): "))
```
